# Setup

```python
# Importar biblioteca
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Criar Objeto 
Vader = SentimentIntensityAnalyzer()
```

Vader é um algoritmo de processamento de sentimentos pré-treinado para dados obtidos a partir de redes sociais. Sua função polarity_scores aceita uma string (Texto) e analisa as probabilidades desta ser positva, negativa ou neutra. Além disto, fornece um resultado ponderado (compound), que vai de -1 até +1, sendo valores > 0 sentimentos positivos e valores < 0 sentimentos negativos

```python
# Teste inicial
Vader.polarity_scores("The batman was a good movie")
```

O vader percebe que a frase acima provavelmente é positiva. Ele também é capaz de distinguir entre a intensidade dos sentimentos dependendo das palavras utilizadas

```python
# Segundo teste, com mais intensidade
Vader.polarity_scores("The batman was an awesome movie!")
```

A utilização de "Awesome" ao invés de "good" e a adição de uma exclamação deixam o algoritmo mais confiante de que a frase é positiva, além de conferir resultado agregado maior que o primeiro teste pelo aumento da intensidade do sentimento positivo. O VADER utiliza um approach "Léxico" no qual as palavras tem seus valores de positividade e negatividade pré-definidos.

Além de palavras, ele é capaz de distinguir emoctions a partir de dígitos normais:

```python
# Carinha feliz
Vader.polarity_scores(":)")
```

```python
# Carinha Triste
Vader.polarity_scores(":(")
```

O vader também consegue distinguir entre emoctions, dado o código em Unicode destes:

```python
# Este é o código Unicode para o Emoji de um rosto sorrindo
Vader.polarity_scores("\U0001F604")
```
